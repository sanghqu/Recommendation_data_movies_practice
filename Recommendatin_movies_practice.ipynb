```python
# recomendation model
def predict_score():
    
    name = input('Enter a movie title: ')
    
    import pandas as pd
    import numpy as np
    import json
    import warnings
    warnings.filterwarnings('ignore')
    from scipy import spatial
    import operator
    
    
    def Similarity(movieId1, movieId2):
        a = movies.iloc[movieId1]
        b = movies.iloc[movieId2]

        genresA = a['genres']
        genresB = b['genres']

        genreDistance = spatial.distance.cosine(genresA, genresB)

        scoreA = a['cast']
        scoreB = b['cast']
        scoreDistance = spatial.distance.cosine(scoreA, scoreB)

        directA = a['crew']
        directB = b['crew']
        directDistance = spatial.distance.cosine(directA, directB)

        wordsA = a['keywords']
        wordsB = b['keywords']
        wordsDistance = spatial.distance.cosine(directA, directB)
        dis=genreDistance + directDistance + scoreDistance + wordsDistance
        return(round(dis,3))
    
    
    
    
    new_movie = movies[movies['original_title'].str.contains(name)].iloc[0].to_frame().T
    def getNeighbors(baseMovie, K):
        distances = []
        for index, movie in movies.iterrows():
            if movie['id'] != baseMovie['id'].values[0]:
                dist = Similarity(baseMovie['id'].values[0], movie['id'])
                distances.append((movie['id'], dist))    
        distances.sort(key=operator.itemgetter(1))
        neighbors = []      
        for x in range(K):
            neighbors.append(distances[x])
        return neighbors
    K = 3
    neighbors = getNeighbors(new_movie, K)
    for neighbor in neighbors:
        s=movies.iloc[neighbor[0]][1]
        print(s)
```

```python
movies['genres'] = movies['genres'].str.strip('[]').str.replace(' ','').str.replace("'",'')
movies['genres'] = movies['genres'].str.split(',')
```

```python
movies['cast'] = movies['cast'].str.strip('[]').str.replace(' ','').str.replace("'",'').str.replace('"','')
movies['cast'] = movies['cast'].str.split(',')
```

```python
def space(col):
  if col is None:
    return ' '
  return str(col)
```

```python
movies['director'] = movies['director'].apply(space)
```

```python
def features(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])

movies['combine_feature'] = movies.apply(features,axis=1)
```

```python
movies[['original_title','combine_feature']]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
cv = CountVectorizer(stop_words='english') 
count_matrix = cv.fit_transform(movies["combine_feature"]) 
cosine_sin = cosine_similarity(count_matrix,count_matrix)
movies = movies.reset_index()
indices = pd.Series(movies.index, index = movies['original_title'])
```

```python
def movies_recommendations(title, cosine_sim=cosine_sin):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    return(pd.DataFrame({'Movie':list(movies.loc[movie_indices,'original_title'])}))
```
